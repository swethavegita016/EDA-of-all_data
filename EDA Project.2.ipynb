# Which city had the highest sales overall?

```python
### Here we need to find the city that had highest sales in 2019. We need to add the city and sales column to find out the
### highest sales in city.From the 'Purchase Adress' column we are splitting the city name by using apply and lamda function.
### apply() function allows us to apply arbitrary function on an entire column.
### lamda X allows us to grab the cell contents.
### we are splitting the address column by ',' and giving index number.as city index number is 1.
df1_fil['City'] = df1_fil['Purchase Address'].apply(lambda x: x.split(',')[1])
df1_fil.head()
```

```python
### Now the dataset has city column in it.
```

```python
### We are adding the sales column by multiplying the quantity ordered and price each of the product.
df1_fil['Sales'] = df1_fil['Quantity Ordered'] * df1_fil['Price Each']
df1_fil.head()
```

```python
### Now the dataset has city and sales column in it.
```

```python
### We are grouping the city by using the groupby function.
### is used for grouping the data according to the categories and apply a function to the categories.
###It also helps to aggregate data efficiently.
city_sales=df1_fil.groupby('City').sum()
city_sales
```

```python
### Plotting the bar graph of sales in cities.
### We are doing the list comprehension for cities to make x and Y are in same order.otherwise the X order will change.
cities=[City for City,df in df1_fil.groupby('City')]
plt.figure(figsize=(10,6))
plt.bar(cities,city_sales['Sales'])
plt.ylabel('Sales')
plt.xticks(rotation='vertical',fontsize = '12')
plt.show()
```

```python
### By looking at the bar graph we can see that san Francisco had highest sales comparing to other cities.
### Maybe san Francisco has good advertising and people may use a lot of electronic gadgets there.
```

```python
### we are plotting the graph of time of the day that has the most sales
### We are doing the list comprehension for hours to make x and Y are in same order.otherwise the X order will change.

hours = [hour for hour, df in df1_fil.groupby('Hour')]
plt.figure(figsize = (10,8))
plt.plot(hours, df1_fil.groupby(['Hour']).count())
plt.xticks(hours)
plt.xlabel('Time (24 hr format)')
plt.ylabel('Number of Orders')
plt.show()
```

# Find out the pair of product that are most sold together.

```python
### Here we need to find the pair of products that has been most sold together.
### We are checking for duplicates by using duplicate function and keeping it as false.This will give us the all duplicated id
### that sold together.
df = df1_fil[df1_fil['Order ID'].duplicated(keep=False)]

### Here we are grouping the order id and product by using the groupby function.
### joining the groupby order id and products by using the transform and lamda x.
### Transform function returns a self-produced dataframe with transformed values 
### after applying the function specified in its parameter.
### lamda X allows us to grab the cell contents.
### dropping the duplicates in order id and grouped.
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df = df[['Order ID', 'Grouped']].drop_duplicates()
df.head(10)
```

```python
### We can see the first 10 rows of products that has been sold together.
```

```python
from IPython.display import display_html
def restartkernel() :
    display_html("<script>Jupyter.notebook.kernel.restart()</script>",raw=True)

### Now we need to iterate these and count in pairs that most sold together often.
### Here we are importing the combinations from itertools and Counter from collections.
from itertools import combinations
from collections import Counter
count=Counter()
### We are selecting the 'Grouped' and splitting by ','
for row in df['Grouped']:
    row_list = row.split(',')
    ##giving the combinations list upto 2
    count.update(Counter(combinations(row_list,2)))
    ###giving the key value count of most common products sold together upto 10
for key,value in count.most_common(10):
    print(key,value)
```
